```python
import numpy as np
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
## Import advertisting dataset
auto = pd.read_csv('..\..\datasets\Auto 2.csv', index_col='Unnamed: 0').reset_index(drop=True)
display(auto)
```

```python
## Perform Linear Regression with foldsCV

X = np.array(auto.horsepower).reshape(-1,1)
y = auto.mpg

mse_list = []

folds = KFold(n_splits=5)
# print(f'{folds.get_n_splits(X)} sets\n')

for i, (train_index, test_index) in enumerate(folds.split(X)):
    # print(f"Fold {i}:")
    # print(f"  Train: index={train_index}")
    # print(f"  Test:  index={test_index}")

    reg = LinearRegression().fit(X[train_index], y[train_index])

    test_preds = reg.predict(X[test_index])

    # print(f'Training MSE: {mean_squared_error(y[test_index], test_preds)}')

    mse_list.append(mean_squared_error(y[test_index], test_preds))

print(f'{folds.get_n_splits(X)} CV Average Test MSE: {round(sum(mse_list)/len(mse_list), 2)}')
```

```python
## Perform Polynomial Regression with LOOCV

X = np.array(auto.horsepower).reshape(-1,1)
y = auto.mpg

poly = PolynomialFeatures(2)
X = poly.fit_transform(X)

mse_list = []

folds = KFold(n_splits=5)
# print(f'{folds.get_n_splits(X)} sets\n')

for i, (train_index, test_index) in enumerate(folds.split(X)):
    # print(f"Fold {i}:")
    # print(f"  Train: index={train_index}")
    # print(f"  Test:  index={test_index}")

    reg = LinearRegression().fit(X[train_index], y[train_index])

    test_preds = reg.predict(X[test_index])

    # print(f'Training MSE: {mean_squared_error(y[test_index], test_preds)}')

    mse_list.append(mean_squared_error(y[test_index], test_preds))

print(f'{folds.get_n_splits(X)} CV Average Test MSE: {round(sum(mse_list)/len(mse_list), 2)}')
```
