```python
import numpy as np
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
## Import advertisting dataset
auto = pd.read_csv('..\..\datasets\Auto 2.csv', index_col='Unnamed: 0')
display(auto)
```

```python
## Will use validation set approach
## Testing linear models using sklearn
## Creation of test and training sets with sklearn.train_test_split of 50%

## Doesn't like when you pass in a series of only one feature. Pushing to numpy array and reshaping
X = np.array(auto.horsepower).reshape(-1,1)
y = auto.mpg

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

reg = LinearRegression().fit(X_train, y_train)

train_preds = reg.predict(X_train)

print(f'Training MSE: {mean_squared_error(y_train, train_preds)}')

test_preds = reg.predict(X_test)

print(f'Training MSE: {mean_squared_error(y_test, test_preds)}')
```

```python
## Perform the same thing with Quadratic Regression

## Doesn't like when you pass in a series of only one feature. Pushing to numpy array and reshaping
X = np.array(auto.horsepower).reshape(-1,1)
y = auto.mpg

## Add Polynomial Features
poly = PolynomialFeatures(2)
X = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

reg = LinearRegression().fit(X_train, y_train)

train_preds = reg.predict(X_train)

print(f'Training MSE: {mean_squared_error(y_train, train_preds)}')

test_preds = reg.predict(X_test)

print(f'Training MSE: {mean_squared_error(y_test, test_preds)}')
```
