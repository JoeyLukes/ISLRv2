```python
import numpy as np
import pandas as pd

import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm
```

```python
## Import Stock Market Data

## Import advertisting dataset
bikeshare = pd.read_csv('..\..\datasets\Bikeshare_hour.csv', index_col='instant')
display(bikeshare)
```

```python
bikeshare[['yr', 'mnth', 'hr', 'weekday', 'workingday', 'holiday']] = bikeshare[['yr', 'mnth', 'hr', 'weekday', 'workingday', 'holiday']].astype('str')
bikeshare.drop(columns=['dteday', 'season', 'yr', 'holiday', 'weekday', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered'], inplace=True)
bikeshare.dtypes
```

```python
exog = bikeshare.drop(columns=['cnt'])
exog = pd.get_dummies(exog, columns=['mnth', 'hr', 'workingday'])
endog = bikeshare.cnt

exog = sm.add_constant(exog)

poisson_model = sm.GLM(endog, exog, faimly=sm.families.Poisson)

poisson_results = poisson_model.fit()

print(poisson_results.summary())
```
